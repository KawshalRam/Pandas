```python
import pandas as pd
df=pd.DataFrame({"city":["new york","chicago","san francisco"],"temperature":[21,14,35]})
df
```

```python
df1=pd.DataFrame({"city":["chicago","new york","orlando","baltimore"],"humidity":[65,68,75,60]})
df1
```

```python
df2=pd.merge(df,df1,on="city",how="inner",indicator=True)
```

It only shows intersection of both the dataframe

basically it is an inner join

```python
df2
```

```python
df3=pd.merge(df,df1,on="city",how="outer",indicator=True)
df3
```

```python
df4=pd.merge(df,df1,on="city",how="right")
df4
```

```python
df5=pd.merge(df,df1,on="city",how="left")
df5
```

```python
df1=pd.DataFrame({"city":["new york","chicago","orlando","baltimore"],"temperature":[21,14,35,38],"humidity":[65,68,71,75]})
df1
```

```python
df2=pd.DataFrame({"city":["chicago","new york","san diego"],"temperature":[21,14,35],"humidity":[65,68,71]})
```

```python
df2
```

```python
df3=pd.merge(df1,df2,on="city",suffixes=("_left","_right"))
df3
```
