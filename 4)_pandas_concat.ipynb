```python
import pandas as pd
import numpy as np
```

```python
india_weather=pd.DataFrame({
    "city":["mumbai","delhi","bangalore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
india_weather
```

```python
us_weather=pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35],
    "humidity":[68,65,75]
})
us_weather
```

```python
df=pd.concat([india_weather,us_weather],keys=['india','us'])
```

```python
df
```

```python
ndf=pd.concat([india_weather,us_weather],ignore_index=True)
```

```python
ndf
```

```python
df.loc['india']
```

```python
df.loc["us"]
```

```python
cdf=pd.concat([india_weather,us_weather],axis=1)
```

```python
cdf
```

```python
xdf=pd.concat([india_weather,us_weather],axis=1,keys=['india','us'])
```

```python
xdf
```

```python
s=pd.Series(["rain","dry","humid"],name="event",index=[2,1,0])
s
```
