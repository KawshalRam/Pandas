#Pivot allows to transform and reshape data


```python
import pandas as pd
```

```python
df=pd.read_csv('/content/weather.csv')
```

```python
df
```

```python
df1=df.pivot(index='date',columns='city')
```

```python
df1
```

```python
df2=df.pivot(index='date',columns='city',values='humidity')
```

```python
df2
```

```python
df3=df.pivot(index='date',columns='city',values='temperature')
df3
```

```python
df4=pd.read_csv('/content/sample_data/weather2.csv')
df4
```

```python
df5=df4.set_index('date')
```

```python
df5
```

```python
df4
```

```python
df6=df4.pivot_table(index='date',columns='city',aggfunc="mean")
df6
```

```python
df7=df4.pivot_table(index='date',columns='city',margins=True)
df7
```
